# Step 3: Statistical Analysis
# We perform deeper statistical analysis like correlations and cross-tabulations.


```python
import sys
sys.path.append("../src")

from utils import load_data, clean_data
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load + clean
df = clean_data(load_data())

# Correlation heatmap
numeric_cols = df.select_dtypes(include='number')
sns.heatmap(numeric_cols.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# Crosstab example: Violation type vs Gender
display(pd.crosstab(df['Violation_Type'], df['Driver_Gender']).head(10))

# Average fine by gender
print("Average fine by gender:\n", df.groupby("Driver_Gender")["Fine_Amount"].mean())
```
